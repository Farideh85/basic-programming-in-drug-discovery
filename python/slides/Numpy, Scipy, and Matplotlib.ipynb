# Numpy

* Library for numeric computations

# Numpy arrays

Python lists do not support mathematical operations:

```python
[1,2,3]+[1,2,3]
```

Numpy arrays support math; try the following:

```python
import numpy as np

a = np.array([1,2,3,4,5])

print(a+a)
print(a-a)
print(a*a)
print(a/a)
```

# Intermezzo - importing libraries

* Importing the full library:
  * **good**: `import numpy as np`; and use as: `np.array`
  * **bad**: `from numpy import *`; and use as: `array`
  * **ugly**: `import numpy`; and use as: `numpy.array`
* Import specific functionality
  * `from library import module.submodule.function`

# Numpy arrays - basic syntax similar to Python lists

```python
a = np.array([1,2,3,4,5])
print('a has {} elements'.format(len(a)))
print('the first element of a is {}'.format(a[0]))
print('the last element of a is {}'.format(a[-1]))  
```

# Numpy arrays - more functionality


```python
a = np.array([1,2,3,4,5])
print(a.min(),a.max(),a.mean(),a.std(),a.sum())
```

Alternative syntax:

```python
print(np.min(a),np.max(a),np.mean(a),np.std(a),np.sum(a))
```

# Multi-dimensional Numpy arrays

Accessing elements:
* `a[i,j]`: element on row i, in column j
* `a[:,j]`: j-th column
* `a[i,:]`: i-th row

# Data visualization - annotation

```python
x = np.linspace(0,10,11)
y = np.exp(x)
# create figure
fig = plt.figure(figsize=(6,4))
# get current axis
ax = plt.gca()
ax.plot(x,y)
# annotate
ax.set(title='a plot',xlabel='x-axis',ylabel='y-axis'); 
```

# Data visualization - Heatmap

```python
x = np.linspace(0,1,11)
y = np.linspace(0,1,11)
X, Y = np.meshgrid(x, y)
z = X*X+Y*Y
plt.imshow(z,cmap='magma')
cb = plt.colorbar();
cb.set_label('x+y')
```
